## Model evaluation

```python
def highlight_max(data, color='bold'):
    '''
    highlight the maximum in a Series or DataFrame
    '''
    attr = 'font-weight: {}'.format(color)
    if data.ndim == 1:  
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)
```

```python
def compare_difficulty_cases(opt,LexSim,distance_metric = 'js-div',append_map=True,split_by='median',report_best=True,report_my=True,report_compare=True,report_obs_number=False,split='test'):
    VM_path = False
    if opt['dataset']=='Semeval':
        if opt['tasks'] == ['A']:
            best = ['KeLP','Beihang-MSRA','IIT-UHH','ECNU','bunji','EICA','SwissAlps','FuRongWang','FA3L','SnowMan']
        elif opt['tasks'] == ['B']:
            best = ['SimBow','LearningToQuestion','KeLP','Talla','Beihang-MSRA','NLM_NIH','UINSUSKA-TiTech','IIT-UHH','SCIR-QA','FA3L']
        elif opt['tasks'] == ['C']:
            best = ['IIT-UHH','bunji','KeLP','EICA','FuRongWang']
    tables = []
    my =[]
    if report_compare:
        compare = ['truly_random']
    else:
        compare = []
    if not report_my:
        my = []
    if not report_best:
        best = []
    models = best+my+compare
    maps = []
    print(models)
    for m in models:
        opt['id']=m
        if m in my:
            VM_path = True
            part = split
        else:
            part = 'primary'
        if append_map:
            maps.append(get_model_MAP(m,VM_path))
        split_df = load_subset_pred_overlap(LexSim,distance_metric,split_by,opt, part, VM_path)
        report_obs = report_obs_number and (m == models[-1])
        result_df = report_results(split_df,opt,report_obs_number=report_obs,max_char=None,rename_model_ids=False,VM_path=VM_path)
        tables.append(result_df)
    print('{} {}'.format(opt['dataset'],opt['tasks'][0]))
    final_table = pd.concat(tables,1)
    if append_map:
        final_table = final_table.append(pd.Series(maps,name='total_map', index=final_table.columns),ignore_index=False)
    return final_table
```

### SemEval systems

```python
opt = {'dataset': 'Semeval', 'datapath': 'data/',
       'tasks': ['A'],'n_gram_embd':False,
        'subsets': ['train_large','test2016', 'test2017'],
       'load_ids': False, 'cache':True,'id':None,
       'simple_padding': False, 'padding': True}
metric = 'js-div'
criterion = 'median'
report_best = True
split = 'median'
difficulty_A = compare_difficulty_cases(opt,LexSim,distance_metric=metric,append_map=False,report_best=report_best,report_my=True,report_compare=True,split_by=criterion,report_obs_number=False,split=split)
difficulty_A.style.apply(highlight_max,axis=1)
```
